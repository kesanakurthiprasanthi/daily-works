# binning

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.DataFrame({
    "age":[18,22,25,30,35,40,45,50]
})
data
```

# Equal width

```python
data["age_bin"]=pd.cut(data["age"],bins=4)
data
```

```python
plt.hist(data["age"],bins=4)
plt.xlabel("age")
plt.ylabel("frequency")
plt.title("equal width binning using pd.cut()")
plt.show()
```

```python
plt.hist(data["age"],bins=3)
plt.xlabel("age")
plt.ylabel("frequency")
plt.title("equal width binning using pd.cut()")
plt.show()
```

# equal frequancy binning

```python
data["age_frequancy_bin"]=pd.qcut(data["age"],q=4)
data
```

```python
data["age_frequency"]=pd.qcut(data["age"],q=4)
data
```

```python
plt.hist(data["age"], bins=4)
plt.xlabel("age")
plt.ylabel("Frequency")
plt.title("Equal_Width Binning Using pd.cut()")
plt.show()
```

```python
bin_min=data["age"].groupby(data["age_bin"]).max()
bin_min
```

```python
bin_min=data["age"].groupby(data["age_bin"]).min()
bin_min
```

```python
bin_min=data["age"].groupby(data["age_bin"]).mean()
bin_min
```
