```python
import pandas as pd
data=pd.DataFrame({
    "hour":[1,2,3,4,5,6,7,8,9,20]
})
data
```

```python
data.describe()
```

# IQR Method

```python
q1=data.quantile(0.25)
q3=data.quantile(0.75)
iqr=q3-q1
lb=q1-1.5*iqr
ub=q3+1.5*iqr
outliers=data[(data<(lb))|(data>(ub))]
outliers
```

# z-score method

```python
import pandas as pd
data=pd.DataFrame({
    "hour":[1,2,3,4,5,6,7,8,9,20]
})
data
```

```python
import numpy as np 
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
```

```python
z_scores=np.abs(stats.zscore(data["hour"]))
outliers=data[z_scores>2]
outliers
```

```python
stats.zscore(data["hour"])
```

```python
plt.scatter(range(len(data)),data["hour"])
plt.xlabel("index")
plt.ylabel("hour")
plt.title("scatter plot")
plt.show()
```
